```python
import pandas as pd

df=pd.read_csv('realistic_dataset.csv')
df.info()
```

```python
df.isnull().sum()
```

```python
df.isnull()
df.dropna(inplace=True)
df.isnull()

df['Current_CGPA']=df["Current_CGPA"].round(2)
df['School_Type_Public']=df['School_Type_Public'].astype(int)
df.info()
```

```python
df
```

```python
numeric=["Family_Income_Monthly","Current_CGPA","Student_Motivation","Physical_Verification_Score","Siblings_Count","Parents_Employed"]
df[numeric]
```

```python
from sklearn.preprocessing import StandardScaler
import joblib
scalar=StandardScaler()

df[numeric]=scalar.fit_transform(df[numeric])
joblib.dump(scalar,"scalar_u1.pkl")
df
```

```python
from sklearn.preprocessing import LabelEncoder

encoder=LabelEncoder()
df["Gender"]=encoder.fit_transform(df["Gender"])
joblib.dump(encoder,"encoder_u1.pkl")
df
```
