```python
import pandas as pd

df=pd.read_csv('maatram_training_dataset.csv')
df.info()
```

```python
df
```

```python
numeric=["Family_Income_Monthly","Current_CGPA","Student_Motivation","Physical_Verification_Score","Siblings_Count","Parents_Employed"]
df[numeric]
```

```python
from sklearn.preprocessing import StandardScaler
scalar=StandardScaler()

df[numeric]=scalar.fit_transform(df[numeric])
df
```

```python
from sklearn.preprocessing import LabelEncoder

encoder=LabelEncoder()
df["School_Type"]=encoder.fit_transform(df["School_Type"])
df
```

```python
df.to_csv("maatram_pre_data.csv",index=False)
```
